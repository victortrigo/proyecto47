```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Generar un nuevo conjunto de datos
np.random.seed(0)
tamaño = 2.5 * np.random.randn(100) + 30  # Tamaño en pies cuadrados
precio = 30000 + (tamaño * 8000) + np.random.randn(100) * 5000  # Precio en dólares
```

```python
# Generar datos de ejemplo
# np.random.seed(0)
# tamaño = 2.5 * np.random.randn(100) + 25  # Tamaño en pies cuadrados
# precio = 20000 + (tamaño * 5000) + np.random.randn(100) * 10000  # Pre<cio en dólares
# Crear un DataFrame

tamaño = 2.5 * np.random.randn(100) + 30  # Tamaño en pies cuadrados
precio = 30000 + (tamaño * 8000) + np.random.randn(100) * 5000  # Precio en dólares

df = pd.DataFrame({'Tamaño': tamaño, 'Precio': precio})
df
```

```python
# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
X = df[['Tamaño']]  # Variable independiente
y = df['Precio']    # Variable dependiente
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
plt.scatter(X, y)
```

```python
plt.scatter(X_train, y_train)
```

```python
plt.scatter(X_test, y_test)
```

```python
# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
# Realizar predicciones
y_pred = modelo.predict(X_test)
```

```python
# Visualizar los resultados
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(X_test, y_pred, color='red')#, linewidth=2, label='Predicción')
plt.title('Regresión Lineal Simple')
plt.xlabel('Tamaño (pies cuadrados)')
plt.ylabel('Precio (dólares)')
plt.legend()
plt.show()
```

```python
# Mostrar los coeficientes
r_squared = modelo.score(X_test, y_test)

print(f'Pendiente (slope): {modelo.coef_[0]}')
print(f'Intersección (intercept): {modelo.intercept_}')
print(f'Coeficiente R²: {r_squared}')
```
